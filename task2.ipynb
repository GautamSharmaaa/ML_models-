```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import time
from google.colab import drive
from tensorflow.keras.applications import ResNet50, VGG16, VGG19, InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input as preprocess_resnet
from tensorflow.keras.applications.vgg16 import preprocess_input as preprocess_vgg
from tensorflow.keras.applications.vgg19 import preprocess_input as preprocess_vgg19
from tensorflow.keras.applications.inception_v3 import preprocess_input as preprocess_inception


drive.mount('/content/drive')

image_paths = [
    "/content/drive/MyDrive/DL lab/00025.jpg",
    "/content/drive/MyDrive/DL lab/00036.jpg",
    "/content/drive/MyDrive/DL lab/00040.jpg",
    "/content/drive/MyDrive/DL lab/00041.jpg",
    "/content/drive/MyDrive/DL lab/00059.jpg",
]

def load_and_preprocess_image(img_path, preprocess_fn):
    img = image.load_img(img_path, target_size=(224, 224))  # Resize
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array = preprocess_fn(img_array)  # Apply preprocessing
    return img, img_array

original_images = []
preprocessed_images = {}

for path in image_paths:
    img, img_resnet = load_and_preprocess_image(path, preprocess_resnet)
    _, img_vgg = load_and_preprocess_image(path, preprocess_vgg)
    _, img_vgg19 = load_and_preprocess_image(path, preprocess_vgg19)
    _, img_inception = load_and_preprocess_image(path, preprocess_inception)

    original_images.append(img)

    preprocessed_images.setdefault("ResNet50", []).append(img_resnet)
    preprocessed_images.setdefault("VGG16", []).append(img_vgg)
    preprocessed_images.setdefault("VGG19", []).append(img_vgg19)
    preprocessed_images.setdefault("InceptionV3", []).append(img_inception)

plt.figure(figsize=(10, 5))
for i, img in enumerate(original_images):
    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.axis("off")
plt.suptitle("Original Images")
plt.show()

models_info = {
    "ResNet50": {
        "model": ResNet50(weights="imagenet", include_top=False),
        "layers": ["conv1_conv", "conv2_block3_out", "conv3_block4_out"]
    },
    "VGG16": {
        "model": VGG16(weights="imagenet", include_top=False),
        "layers": ["block1_conv2", "block3_conv3", "block5_conv3"]
    },
    "VGG19": {
        "model": VGG19(weights="imagenet", include_top=False),
        "layers": ["block1_conv2", "block3_conv4", "block5_conv4"]
    },
    "InceptionV3": {
        "model": InceptionV3(weights="imagenet", include_top=False),
        "layers": ["conv2d_2", "mixed2", "mixed7"]
    }
}

for model_name, info in models_info.items():
    base_model = info["model"]
    layer_names = info["layers"]

    feature_extractors = [Model(inputs=base_model.input, outputs=base_model.get_layer(layer).output) for layer in layer_names]

    print(f"\nüîç Extracting features using {model_name}...")

    for i, extractor in enumerate(feature_extractors):
        start_time = time.time()
        plt.figure(figsize=(10, 5))

        for j, img_array in enumerate(preprocessed_images[model_name]):
            feature_maps = extractor.predict(img_array)
            feature_map = np.mean(feature_maps[0], axis=-1)

            plt.subplot(1, 5, j + 1)
            plt.imshow(feature_map, cmap="viridis")
            plt.axis("off")
            plt.title(f"Image {j+1}")

        elapsed_time = time.time() - start_time
        print(f"‚è≥ Time taken for {layer_names[i]}: {elapsed_time:.2f} seconds")

        plt.suptitle(f"{model_name} - Features from Layer: {layer_names[i]}")
        plt.show()
```

```python
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE


comparison_data = []

for model_name, info in models_info.items():
    base_model = info["model"]
    layer_names = info["layers"]

    for i, extractor in enumerate([Model(inputs=base_model.input, outputs=base_model.get_layer(layer).output) for layer in layer_names]):
        start_time = time.time()

        feature_shapes = []
        feature_variances = []
        all_features = []

        for img_array in preprocessed_images[model_name]:
            feature_maps = extractor.predict(img_array)
            feature_map = np.mean(feature_maps[0], axis=-1)

            feature_shapes.append(feature_map.shape)
            feature_variances.append(np.var(feature_map))
            all_features.append(feature_map.flatten())

        elapsed_time = time.time() - start_time

        comparison_data.append({
            "Model": model_name,
            "Layer": layer_names[i],
            "Feature Size": feature_shapes[0],
            "Variance": np.mean(feature_variances),
            "Time Taken (s)": elapsed_time
        })

df_comparison = pd.DataFrame(comparison_data)
print(df_comparison)

plt.figure(figsize=(10, 5))
sns.barplot(x="Model", y="Time Taken (s)", hue="Layer", data=df_comparison)
plt.title("Feature Extraction Time for Each Model & Layer")
plt.ylabel("Time (seconds)")
plt.xlabel("Model")
plt.legend(title="Layer", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x="Model", y="Variance", hue="Layer", data=df_comparison)
plt.title("Feature Variance Across Models & Layers")
plt.ylabel("Variance")
plt.xlabel("Model")
plt.legend(title="Layer", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()

all_features = []
labels = []

for model_name, info in models_info.items():
    base_model = info["model"]
    extractor = Model(inputs=base_model.input, outputs=base_model.get_layer(info["layers"][1]).output)

    for img_array in preprocessed_images[model_name]:
        feature_maps = extractor.predict(img_array)
        feature_map = np.mean(feature_maps[0], axis=-1).flatten()
        all_features.append(feature_map)
        labels.append(model_name)

max_features = max(feature_map.size for feature_map in all_features)

def resize_feature_vector(feature_map, target_size):
    feature_flattened = feature_map.flatten()
    if feature_flattened.size > target_size:
        return feature_flattened[:target_size]
    else:
        return np.pad(feature_flattened, (0, target_size - feature_flattened.size), mode='constant')

all_features = np.array([resize_feature_vector(f, max_features) for f in all_features])

num_components = min(10, all_features.shape[0], all_features.shape[1])
pca = PCA(n_components=num_components).fit_transform(all_features)

tsne = TSNE(n_components=2, perplexity=5, random_state=42)
tsne_features = tsne.fit_transform(pca)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=tsne_features[:, 0], y=tsne_features[:, 1], hue=labels, palette="deep")
plt.title("t-SNE Visualization of Extracted Features")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend(title="Model")
plt.show()
```
